```python
# Already scraped temperature data from tianqihoubao.com.
# Merging all the files into one data frame for clustering now.
import pandas as pd
import os
import datetime as dt
```

```python
high = pd.DataFrame()
low = pd.DataFrame()
for x, y, z in os.walk('data/1103/'):
    for f in z:
        city = f.split('_')[0]
        tmp = pd.read_csv(x + f)[['date', 'high', 'low']].set_index('date')
        tmpHigh = tmp[['high']]
        tmpLow = tmp[['low']]
        tmpHigh.columns = [city]
        tmpLow.columns = [city]
        high = high.merge(tmpHigh, left_index=True, right_index=True, how='outer')
        low = low.merge(tmpLow, left_index=True, right_index=True, how='outer')
```

```python
print(high.shape, low.shape)
```

```python
high = high.dropna(axis=0, how='any')
low = low.dropna(axis=0, how='any')
print(high.shape, low.shape)
```

```python
high.to_csv('1103_cities48_2015_temperature_high.csv', index=True, encoding='utf-8')
low.to_csv('1103_cities48_2015_temperature_low.csv', index=True, encoding='utf-8')
```
